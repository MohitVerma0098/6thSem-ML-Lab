```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([8, 10, 12]).reshape(-1, 1)
y = np.array([10, 13, 16])

x_bar = np.mean(X)
y_bar = np.mean(y)

numerator = np.sum((X - x_bar) * (y.reshape(-1, 1) - y_bar))
denominator = np.sum((X - x_bar)**2)

if denominator == 0:
    b1 = 0
else:
    b1 = numerator / denominator

b0 = y_bar - b1 * x_bar

diameter_20 = np.array([[20]])
predicted_price = b0 + b1 * diameter_20

print("Predicted price for 20-inch pizza: $", round(predicted_price[0,0], 2))
print("\nModel Equation:")
print("Price = {:.2f} * Diameter + {:.2f}".format(b1, b0))

plt.scatter(X, y, color='red', label='Actual Data')
plt.plot(X, b0 + b1 * X, color='blue', label='Regression Line')
plt.scatter(20, predicted_price, color='green', label='Prediction (20 inch)')
plt.xlabel("Diameter (inches)")
plt.ylabel("Price ($)")
plt.title("Linear Regression: Pizza Price Prediction (X-bar, Y-bar method)")
plt.legend()
plt.show()
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Upload dataset (Run this in Colab)
from google.colab import files
uploaded = files.upload()

# Step 3: Load dataset
df = pd.read_csv('canada_per_capita_income.csv')

# Step 4: Display basic info
print("Dataset Info:")
print(df.info())

print("\nDataset Description:")
print(df.describe())

# Step 5: Prepare data
X = df[['year']]
y = df['per capita income (US$)']

# Step 6: Create & Train model
model = LinearRegression()
model.fit(X, y)

# Step 7: Predict for 2020
predicted_income = model.predict([[2020]])
print("\nPredicted Per Capita Income in 2020:")
print(predicted_income[0])

# Step 8: Model parameters
print("\nModel Coefficient:", model.coef_[0])
print("Model Intercept:", model.intercept_)

# Step 9: Plot regression line
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Canada Per Capita Income Prediction")
plt.show()
```

```python
# ==========================================
# MULTIPLE LINEAR REGRESSION - HIRING DATA
# ==========================================

# Step 1: Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Step 2: Upload file (Run in Colab)
from google.colab import files
uploaded = files.upload()

# Step 3: Load dataset
df = pd.read_csv("hiring.csv")

print("Dataset Info:")
print(df.info())

print("\nDataset:")
print(df)

# Step 4: Handle missing values

# Convert experience words (if any) to numbers
# Replace words like 'two', 'three' etc. (if present)
df['experience'] = df['experience'].replace({
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleven': 11,
    'twelve': 12
})

# Fill missing experience with 0
df['experience'] = df['experience'].fillna(0)

# Fill missing test_score with mean
imputer = SimpleImputer(strategy='mean')
df[['test_score(out of 10)']] = imputer.fit_transform(df[['test_score(out of 10)']])

# Step 5: Prepare data
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

# Step 6: Train model
model = LinearRegression()
model.fit(X, y)

# Step 7: Predictions
salary_1 = model.predict([[2, 9, 6]])
salary_2 = model.predict([[12, 10, 10]])

print("\nPredicted Salary for (2 yr, 9 test, 6 interview):")
print(salary_1[0])

print("\nPredicted Salary for (12 yr, 10 test, 10 interview):")
print(salary_2[0])

# Step 8: Model Equation
print("\nModel Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)
```
